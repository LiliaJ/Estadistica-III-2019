```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import pandas as pd
import scipy.integrate as integrate
import scipy.stats as sts
from __future__ import division
%matplotlib inline
```

```python
class EstadisticaIII:
    def __init__(self):
        self.X = None
        self.Y = None
    def sim(self,f,E,n):
        R = []
        Y = [0]+[integrate.quad(f,-np.inf,x)[0] for x in E]
        for i in range(n):
            u = np.random.uniform()
            R.append(max([E[x] for x in range(len(E)) if Y[x]<u]))
        return R
    def VarianzaAsintotica(self,f,p,E,I,n,k):
        L = []
        for i in range(k):
            L.append(m.sqrt(n)*(E(self.sim(f,I,n))-p))
        plt.hist(L)
        return np.var(L)
    def Fn(self,y):
        return len([x for x in self.X if x<=y])/len(self.X)
    def bandadeconfianza(self,a):
        y = np.arange(min(self.X),max(self.X)+0.02,0.01)
        n = len(self.X)
        En = m.sqrt(m.log(2/a)/(2*n))
        L =[max(self.Fn(x)-En,0) for x in np.arange(min(self.X),max(self.X)+0.02,0.01)]
        U =[min(self.Fn(x)+En,1) for x in np.arange(min(self.X),max(self.X)+0.02,0.01)]
        F = [self.Fn(x) for x in y]
        self.LowerBound = lambda x: max(self.Fn(x)-En,0)
        self.UpperBound = lambda x: min(self.Fn(x)+En,1)
        p = plt.plot(y,U,y,L,y,F)
    def KS(self,F):
        Xi = sorted(self.X)
        self.n = len(self.X)
        Dn = []
        for i in range(self.n):
            Dn.append(max((i+1)/self.n-F(Xi[i]),F(Xi[i])-i/self.n))
        self.D = max(Dn)
        K = np.argmax(Dn)
        Femp = [self.Fn(x) for x in np.arange(min(self.X),max(self.X)+0.02,0.01)]
        F0 = [F(x) for x in np.arange(min(self.X),max(self.X)+0.02,0.01)]
        P = plt.plot(np.arange(min(self.X),max(self.X)+0.02,0.01),Femp,
                np.arange(min(self.X),max(self.X)+0.02,0.01),F0,
                Xi[K],F(Xi[K]),'g^')
```

```python
E = EstadisticaIII()
```

```python
E.X = np.random.beta(1,3,1000)
```

```python
E.bandadeconfianza(0.2)
```

```python
E.KS(F=lambda x:sts.beta.cdf(x,1,3))
```
