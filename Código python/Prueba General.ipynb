```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from __future__ import division
%matplotlib inline
```

```python
F = lambda x: 1-m.exp(-x)
```

```python
Fn = lambda y,X:len([x for x in X if x<=y])/len(X)
D = np.random.exponential(1,1000)
```

```python
X = np.arange(min(D),max(D)+0.01,0.01)
Y = [(F(x)-Fn(x,D))**2 for x in X]
```

```python
def f(x):
    r=1
    for i in range(1,x+1):
        print ('addition{}'.format(i))
        r = r+i
    return r
```

```python
f(3)
```

```python
plt.plot(X,Y)
```
